a. Consider the following dataset. Split the data into a 30-70 test-training set

```python
import seaborn as sns
diamonds = sns.load_dataset('diamonds')
data = diamonds.to_numpy()
y = data[:,6]
X = data[:,(0,4,5,7,8,9)] # restrict ourselves to real numbers only

diamonds.head()
```

```python
random.seed(1)

train_indicies, test_indicies = create_test_train_indicies(.7, X.shape[0])

X_train = X[train_indicies,:]
X_test = X[test_indicies,:]
y_train = y[train_indicies]
y_test = y[test_indicies]
```

```python
def MSE(prediction_errors):
  return sum(prediction_errors ** 2) / len(prediction_errors)
```

# Question 2

a. Using the math package write a function to compute $(x_0 + \exp(x_0) + x_1)^3$






```python
def my_func(X):
  return (X[0] + math.exp(X[0])+X[1])**3
```

```python
# test function
import numpy as np
assert my_func(np.array([0, 0])) == 1.0
assert abs(my_func(np.array([0.1, 0.2])) - 2.7745174378) < 1e-5
```

b. Compute the derivative of the function at $x_0 = 1, x_1 = 1$ with respect to $x_1$ using finite differences


```python
def multivar_finite_diff(func, var, h):
  h_point = [var[0], var[1] + h]
  return (func(h_point) - func(var)) / h
```

```python
multivar_finite_diff(my_func, [1,1], h=1e-5)
```

c. Reimplement the function in pytorch


```python
import torch
def my_pytorch_func(x):
  return torch.pow(x[0] + torch.exp(x[0])+x[1],3)
```

```python
# test function
assert my_pytorch_func(torch.tensor([0, 0])) == 1.0
assert abs(my_pytorch_func(torch.tensor([0.1, 0.2])) - 2.7745174378) < 1e-5
```

d. Recompute the derivative from (b) using pytorch


```python
x0_val = torch.ones(1, requires_grad=False) # requires_grad=True means differentiate with respect to this tensor
x0_val
x1_val = torch.ones(1, requires_grad=True)
x1_val

f = my_pytorch_func([x0_val,x1_val]) # forward pass
f
f.backward()
```

a. Write a function to compute $f(x,y) = y^T A x$ where A is defined to be

```python
import random, numpy as np
random.seed(1)
A = np.random.random((5,8)) # creates a randomly generated matrix
def compute_fxy(x, y):
  return np.dot(y, np.matmul(A,x))
```
