b. Write a function that finds the largest element of an array that is less than five. If no such element exists return NaN. Create your own tests to demonstrate it works.

```python
def largest_under_5(arr):
  largest_index = 0

  for i in range(0,len(arr)):

    if arr[i] < 5:
      if (arr[i] >= arr[largest_index]):
        largest_index = i
    else:
      continue
  if (arr[largest_index] >= 5):
    return np.nan

  else:
    return arr[largest_index]

# TESTS
test_arr1 = [0,1,3,2,4,6]
test_arr2 = [5,9,20,13]

largest_under_5(test_arr1)
largest_under_5(test_arr2)
```

c. Write a function to compute the matrix $C_{i,j} = \frac{A_{i,j}}{A_{i,j}^2 + 1}$. It should pass the subsequent tests.

```python
def compute_C(A):
  # write function here
  C = np.zeros((len(A),len(A[0])))
  for i in range(len(A)):
    for j in range(len(A[0])):

      C[i][j] = (A[i][j])/ ((A[i][j])**2+1)

  return C
```

```python
assert (compute_C(np.ones([2,2])) == np.ones([2,2]) / 2).all()
C = compute_C(np.array([[2.0,5.0],[4.0,3.0]]))
assert C[0,1] == 5 / (5 ** 2 + 1)
```

```python
def smallest_cons_diff(arr):
  small_diff = 9999

  for i in range(0,len(arr)-1):
    diff = abs(arr[i]-arr[i+1])
    if diff < small_diff:
      small_diff = abs(arr[i]-arr[i+1])

  return small_diff




# TESTS
test_arr1 = [0,1,3,2,4,6]
#test_arr2 = [5,9,20,13]

smallest_cons_diff(test_arr1)
#smallest_cons_diff(test_arr2)
```

# Question 4

Plot the following function in python: $f(x) = \frac{0.1 + x^2}{x^2 + 1}$ across the range $x = -2 \text{ to } 2$. Put the y-axis on a log scale. Label the axes and give the graph a title of "My first function plot in python".


**Hints:** `fVectorized = np.vectorize(f)` creates a vectorized version of the function `f`, `np.linspace(start, stop, num)` creates a grid starting from start, ending at stop with num grid points.

```python
import matplotlib.pyplot as plt

def f(x):
    return (0.1 + x**2)/(x**2 + 1)

fVectorized = np.vectorize(f)

x = np.linspace(-2, 2, 400)
y = fVectorized(x)

plt.figure(figsize=(8, 6))
plt.plot(x, y)

plt.yscale("log")
plt.xlabel("x")
plt.ylabel("y")
plt.title("My first function plot in python")

plt.show()
```
